```python
#Cerința 1: Determină luna cu cea mai mare rată de renunțări (%) din totalul membrilor.

import pandas as pd
import matplotlib.pyplot as plt

# ---------------------- Datele introduse manual ----------------------

abonamente_df = pd.DataFrame({
    'Luna': ['Ian', 'Feb', 'Mar', 'Apr', 'Mai', 'Iun', 'Iul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Total Membri': [193, 170, 177, 171, 197, 192, 189, 203, 174, 179, 184, 181],
    'Renuntari': [14, 23, 28, 23, 7, 10, 15, 9, 34, 23, 16, 19],
    'Abonamente Noi': [56, 29, 86, 58, 52, 44, 36, 80, 5, 36, 36, 38]
})

echipamente_oradea = pd.DataFrame({
    'Echipament': [
        'Aparat multifunctional', 'Banda alergare', 'Covoras fitness', 'Gantere',
        'Bicicleta statica', 'Covoras fitness', 'Aparat multifunctional',
        'Bicicleta statica', 'Banda alergare', 'Gantere', 'Aparat multifunctional',
        'Gantere', 'Covoras fitness', 'Bicicleta statica', 'Aparat multifunctional',
        'Gantere', 'Covoras fitness', 'Banda alergare', 'Aparat multifunctional',
        'Banda alergare', 'Bicicleta statica', 'Covoras fitness', 'Aparat multifunctional',
        'Gantere'
    ],
    'Valoare Totala (€)': [
        21666, 3972, 17184, 5109, 5209, 8491, 14920, 15005, 2484, 3288,
        5922, 4346, 8419, 2500, 6507, 4032, 21708, 4116, 6270, 654, 18304, 13446, 962, 4194
    ]
})

# -------------------- Cerința 1 --------------------

abonamente_df['Rata Renuntari (%)'] = abonamente_df['Renuntari'] / abonamente_df['Total Membri'] * 100
luna_max = abonamente_df.loc[abonamente_df['Rata Renuntari (%)'].idxmax()]
print(f"Cea mai mare rată de renunțări este în luna {luna_max['Luna']}: {luna_max['Rata Renuntari (%)']:.2f}%")

# -------------------- Cerința 2 --------------------

top3 = echipamente_oradea.groupby('Echipament')['Valoare Totala (€)'].sum().sort_values(ascending=False).head(3)
print("\nTop 3 echipamente Oradea după valoare totală:")
print(top3)

# Reprezentare grafică
top3.plot(kind='bar', title='Top 3 Echipamente Oradea (după valoare totală)', ylabel='Valoare (€)', xlabel='Echipament')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# -------------------- Cerința 2 --------------------

top3 = echipamente_oradea.groupby('Echipament')['Valoare Totala (€)'].sum().sort_values(ascending=False).head(3)
print("\nTop 3 echipamente Oradea după valoare totală:")
print(top3)

# Reprezentare grafică
top3.plot(kind='bar', title='Top 3 Echipamente Oradea (după valoare totală)', ylabel='Valoare (€)', xlabel='Echipament')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
# -------------------- Cerința 3 --------------------

abonamente_df['Trimestru'] = ['Q1'] * 3 + ['Q2'] * 3 + ['Q3'] * 3 + ['Q4'] * 3
trimestrial = abonamente_df.groupby('Trimestru')[['Abonamente Noi', 'Renuntari']].sum()
print("\nComparare pe trimestre:")
print(trimestrial)

trimestrial.plot(kind='bar', title='Abonamente Noi vs Renunțări pe Trimestre', ylabel='Număr clienți')
plt.tight_layout()
plt.grid(True)
plt.show()
```
